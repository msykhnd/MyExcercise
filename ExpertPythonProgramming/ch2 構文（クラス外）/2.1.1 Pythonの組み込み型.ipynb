```python
import time

substrings = "新型コロナ"*10000
s = time.time()
string = ""
for substring in substrings:
    string += substring
e = time.time()
print(e-s)
print(string)

```

```python
import time

substrings = "新型コロナ"*10000
s = time.time()
string = "!".join(substrings)

e = time.time()
print(e-s)
print(string)
```
